```python
from my_utils.base_data_handler import BaseDataHandler #type: ignore
```

```python
handler = BaseDataHandler('data/processed/tracks_processed_normalized.csv')
group = handler.df.sort_values('popularity', ascending=False)
mask = group['popularity'] > 25 # top 75% songs
group = group[mask]
group.dropna(inplace=True)
group[['popularity', 'track_name', 'artists', 'duration_ms']]
```

```python
handler.try_update_og_df()
handler.df = group.copy()
```

```python
group_by_artist = handler.df.groupby('artists').sample(n=1).reset_index(drop=True)
group_by_artist.info()
```

```python
group_by_genre = group_by_artist.groupby('track_genre').sample(n=2, replace=False).reset_index(drop=True)
group_by_genre.info()
```
